```python
import pandas as pd
```

```python
playstore = pd.read_csv("playstore.csv")
playstore = playstore.drop_duplicates(subset="App").reset_index(drop=True)
playstore.columns = playstore.columns.str.lower().str.replace(' ', '_')
playstore["price"] = playstore["price"].str.replace("$", "", regex=False).astype(float)
```

```python
playstore
```

```python
playstore["mean_price"] = playstore["price"].round(2)
playstore["mean_rating"] = playstore["rating"].round(2)
playstore["mean_reviews"] = playstore["reviews"].round(1)
```

```python
pivot_table = playstore.pivot_table(values=["mean_price", "mean_rating", "mean_reviews"], index=["category", "type"], aggfunc="mean")
```

```python
pivot_table
```

```python
pivot_table.to_csv("res.csv", sep=",")
```

```python
playstore = pd.read_csv("playstore.csv")
playstore = playstore.drop_duplicates(subset="App").reset_index(drop=True)
playstore['Price'] = pd.to_numeric(playstore['Price'].str.replace('$', ''), errors='coerce')
result = playstore.groupby(['Category', 'Type']).agg({'Price':'mean', 'Rating':'mean', 'Reviews':'mean'})
result.columns = result.columns.str.replace('Price', 'mean_price')
result.columns = result.columns.str.replace('Rating', 'mean_rating')
result.columns = result.columns.str.replace('Reviews', 'mean_reviews')
result['mean_price'] = result['mean_price'].round(2)
result['mean_reviews'] = result['mean_reviews'].round(2)
result['mean_rating'] = result['mean_rating'].round(1)
result.to_csv("resultat.csv")
```
