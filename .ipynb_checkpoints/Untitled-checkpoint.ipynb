```python
import pandas as pd
```

```python
playstore = pd.read_csv("playstore.csv")
```

```python
a = playstore.drop_duplicates()
```

```python
a["mean_price"] = a["price"]
a["mean_rating"] = a["rating"]
a["mean_reviews"] = a["reviews"]
```

```python
pivot_table = a.pivot_table(values=["mean_price", "mean_rating", "mean_reviews"], index=["category", "type"], aggfunc="mean").round(2)
pivot_table["mean_reviews"] = pivot_table["mean_reviews"].round(1)
```

```python
import pandas as pd

# Создание срезов данных
df1 = playstore.iloc[:3]
df2 = playstore.iloc[5:8]
df3 = playstore.iloc[15:19]

# Выбор нужных колонок
columns = ["App", "Size", "Genres", "Current Ver"]
df1 = df1[columns]
df2 = df2[columns]
df3 = df3[columns]

# Объединение датафреймов
final_df = pd.concat([df1, df2, df3])

# Сохранение в формате CSV
final_df.to_csv("final_dataframe.csv", index=True)
```
