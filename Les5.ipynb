```python
# Преобразование столбцов dateAdded и dateUpdated из строк в datetime
data['dateAdded'] = pd.to_datetime(data['dateAdded'])
data['dateUpdated'] = pd.to_datetime(data['dateUpdated'])

# Вычисление разницы в днях и добавление столбца update_delta
data['update_delta'] = (data['dateUpdated'] - data['dateAdded']).dt.days

# Группировка по городам и расчет среднего значения update_delta и максимальной широты
result = data.groupby('city', as_index=False).agg({'update_delta': 'mean', 'latitude': 'max'})

# Сохранение среднего значения update_delta для Zephyrhills в переменную zep_mean
zep_mean = result[result['city'] == 'Zephyrhills']['update_delta'].iloc[0]
```
